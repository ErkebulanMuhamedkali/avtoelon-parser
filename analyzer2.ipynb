```python
newDff = advertsDF[['price', 'Город']] #newDff = new_df[['price', 'Город']] если только по городам
avgValue = newDff.groupby(['Город']).mean()
avgValue.sort_values('price', ascending=False)
```

```python
brandPriceDF = advertsDF[['brand', 'price']]
avgValue = brandPriceDF.groupby(['brand']).mean()
avgValue = avgValue.sort_values('price', ascending=False)
avgValue


chevrolet = brandPriceDF[brandPriceDF['brand'] == 'Chevrolet']
print(f'chevrolet: {chevrolet.mean()}')


lada = brandPriceDF[brandPriceDF['brand'].str.contains('Lada')]
print(f'lada: {lada.mean()}')

daewoo = brandPriceDF[brandPriceDF['brand'] == 'Daewoo']
print(f'daewoo: {daewoo.mean()}')
```
